```python
import numpy as np
from scipy import spatial
```

Create word embedding vectors and save them into a dict using the pre-trained embeddings.

```python
embeddings_dict = {}

with open("dolma_300_2024_1.2M.100_combined.txt", 'r', encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
```

Define a method to find nearest words using spatial from Scipy.

```python
def find_closest_embeddings(embedding):
    return sorted(embeddings_dict.keys(), key=lambda word: spatial.distance.euclidean(embeddings_dict[word], embedding))
```
