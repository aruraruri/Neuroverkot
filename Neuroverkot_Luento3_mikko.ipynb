```python
import numpy as np
```

```python
def ReLU(x):
    return np.maximum(0, x)

def softmax(x):
    y = np.exp(x)
    return y / np.sum(y, axis=0)
```

```python
X = np.array([[2, 5, -3]]).T
W1 = np.array([[2.0, -1.2, 0.7],
               [-3.0, 1.0, -2.3]])
W2 = np.array([[1.5, 2.0],
                [-2.0, 1.0]])
```

```python
Z1 = W1 @ X
A1 = ReLU(Z1)
Z2 = W2 @ A1
A2 = softmax(Z2)

print(A2)
```
